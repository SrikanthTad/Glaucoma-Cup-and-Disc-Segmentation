```python
import tensorflow as tf
import numpy as np
import matplotlib as plt
import cv2
import os
```

```python
folder = '/data2/SrikanthData/ResearchWork2/tensorflow/Project/img_data/REFUGE/test/test_mask/'
for filename in os.listdir('/data2/SrikanthData/ResearchWork2/tensorflow/Project/img_data/REFUGE/test/test_mask/'):
    img = cv2.imread(os.path.join(folder,filename),cv2.IMREAD_UNCHANGED)
    print(os.path.join(folder,filename))
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            # Access the current pixel values (R, G, B)
            v = img[y, x]
            if v == 65535:
                v = 32768
                img[y,x] = v
            
            
    cv2.imwrite('/data2/SrikanthData/ResearchWork2/tensorflow/Project/img_data/REFUGE/test_disc/' + filename, img)
```
